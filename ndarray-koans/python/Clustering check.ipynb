```python
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
import pandas as pd

alt.renderers.enable('notebook')
```

```python
dataset = np.load("../clustered_dataset.npy")
clusters = np.load("../clustered_memberships.npy")
```

```python
source = pd.DataFrame({"x": dataset[:, 0], "y": dataset[:, 1], "cluster_index": clusters})

alt.Chart(source).mark_circle(size=60).encode(
    x='x',
    y='y',
    color='cluster_index:N'
).interactive()
```
